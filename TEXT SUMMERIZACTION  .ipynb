```python
nlp=spacy.load('en_core_web_sm')
nlp# LOAD ENG LANG IN MODEL 
```

```python
doc=nlp(text)
doc# AUTOMATICALLY BREAKS THE LINE WHERE \n with FULLSTOP
```

```python
tokens=[token.text for token in doc]# this are the rext present here
tokens# we tokenize the text with the help of list comprehension but we can see that there are many unwanted words in doc
```

```python
from string import punctuation
```

```python
print(punctuation)
```

```python
STOP_WORDS# IS,THE,AND
```

```python
STOP_WORDS=list[STOP_WORDS]
STOP_WORDS
```

```python
word_frequency={}
word_frequency # empty dict
```

```python
for i in doc:
    if i.text.lower() not in STOP_WORDS:
        if i.text.lower() not in punctuation:
            if i.text not in word_frequency.keys():# for keys
                word_frequency[i.text]=1
            else:
                word_frequency[i.text]+=1
#i.text represents the text content of the current word
```

```python
word_frequency # it provides the keywords:frequency
# without STOPWORDS and PUNCTUATION
```

```python
max_frequency = max(word_frequency.values())
max_frequency
# lets search for most repeated words
```

```python
# normalization
for i in word_frequency.keys():
    word_frequency[i]=word_frequency[i]/max_frequency#37
print(word_frequency)
```

```python
sent_token=[sent for sent in doc.sents]
sent_token
```

```python
sent_frequency={}
for sent in sent_token:
    for word in sent:  # Iterate through words in the sentence
        if word.text.lower() in word_frequency.keys():  # Check if the word is in word_frequency
            if sent not in sent_frequency.keys():  # Check if the sentence is not in sent_frequency
                sent_frequency[sent] = word_frequency[word.text.lower()]  # Assign word frequency to sentence
            else:
                sent_frequency[sent]+=word_frequency[word.text.lower()]
sent_frequency
```

```python
from heapq import nlargest
```

```python
len(sent_token)
```

```python
required_length=int(len(sent_token)*0.4)# we will get a float value so we used int
required_length
```

```python
#we have to select maximum 4 sentences out of all sentences 
summary = nlargest(required_length,sent_frequency, key = sent_frequency.get)
```
